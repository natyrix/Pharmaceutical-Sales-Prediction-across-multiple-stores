#### Import Modules

```python
from sqlalchemy import create_engine
import pandas as pd
import sys, os
sys.path.append(os.path.abspath(os.path.join('../scripts')))
from read_write_util import ReadWriteUtil
```

```python
reader = ReadWriteUtil()
df_train = reader.dvc_get_data("../data/train.csv", 'v2')
df_train.head()
```

```python
df_train.shape
```

#### Saving Data to MYSQL Database 

```python
df = df_train[:5000]
df.shape
```

```python
#sales_prediction
sql_engine = create_engine('mysql+pymysql://root:@localhost/sales_prediction')
```

```python
try:
    print('writing to the database')
    frame = df.to_sql(
        "sales_prediction_training", con=sql_engine, if_exists='replace')
    print('successful')
except Exception as e:
  print("Error writing to database: ", e)
```

```python
pd.read_sql("select * from sales_prediction.sales_prediction_training", sql_engine)
```
